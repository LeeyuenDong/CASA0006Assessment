####  The Pearson correlation coefficient 
The heat map is drawed based on the value of the Pearson correlation coefficient between each pair of variables. If the value is positve, the color is red, otherwise, it's blue. Besides, it's value is between -1 and 1. The closer its absolute value close to 1 the deeper the color is.
 $$ r_{xy}= \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum(x_i - \bar{x})^2{\sum(y_i - \bar{y})^2}}}$$

- $r_{xy}$ is the Pearson correlation coefficient between observed $X$ and $Y$
- $\bar{x},\bar{y}$ is the average of $X$ and $Y$

Overall, the variables appear to be relatively well correlated, and it is difficult to directly identify which variables show strong correlation. 

Therefore we will introduce the VIF method to identify the variables again and check which variables will be dropped.

The mathematic explanation of VIF is:
$$ VIF_i=\frac{1}{1-R^2_i} $$

```python
# calculating VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(data, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: (default 5) the threshould VIF value. If the VIF of a variable is greater than thresh, it should be removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data
        data_with_const = add_constant(data)

        vif_data = pd.Series([variance_inflation_factor(data_with_const.values, i) 
               for i in range(data_with_const.shape[1])], name= "VIF",
              index=data_with_const.columns).to_frame()

        # drop the const
        vif_data = vif_data.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        # If there are multiple variabels with VIF>thresh, only one of them is removed. This is because we want to keep as many variables as possible
        if vif_data.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_data.index[vif_data.VIF == vif_data.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            data = data.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return data
```

```python
data_predictors_selected_VIF = drop_column_using_vif_(df1.drop('lnprice', axis=1))
# Note that there is a 'while' loop in the drop_column_using_vif_. In each iteration, one variable is dropped.
print("The columns remaining after VIF selection are:")
print(data_predictors_selected_VIF.columns)
```

The VIF testing help us to select the variables and we will consider both correaltion matrix and VIF testing outcomes to decide which variables will be enrolled in our model.

## Machine Learning Model Building

In the model buliding part we will use the test accuracy ($R^2$) and MSE (Mean Squared Error) as evaluation metrics to measure how good a model is and compare training and testing outcomes to determine if the model is overfitting.


* $R^2$ (Coefficient of the determination)

$R^2$ measures the percentage of variance of the responde variable explained by the model. It's a value between 0 and 1. If the value closer to 1 indicates a better prediction, otherwise the it has a bad performance.

The mathematical expression is

$$ \operatorname{R^2} = \frac{SSR}{SST} $$
Sum of the Squared Regression(SSR) is the residual that can be explained by the regression line. 
 $$ SSR = \sum_{i=1}^n \left(\hat{y_i} - \bar{y}\right)^2 $$
Total Sum of the Squared Error(SST) is the total residual. 
 $$SST = SSR + SSE$$
 $$ SSE = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$


* MSE(Mean Squared Error)

The mathematical expression is

$$\operatorname{MSE}= \frac{1}{n}\sum_{i=1}^n(\hat{y_i} - y_i)^2$$



* $y_i$ = Each observed data point. 
* $\hat{y_i}$ = Predicted data point for each $x_i$ depending on i.

Sigmoid

Hyperboic tangent (tanh)
$
a(z) = \tanh(z)

Rectified linear unit (Relu)
$
a(z) = \max(0, z)

# K-fold Cross Validation

```python
def cv_to_choose_parameter(parameters):
    '''
    param: parameters is a dictory, it represent the parameter and their values we want to choose
    description:use grid research to find the parameter value which has the smallest MSE and output the best parameter and the corresponding score
    '''
    rf = RandomForestRegressor()
    clf = model_selection.GridSearchCV(rf, parameters)
    clf.fit(X_train, y_train)
    means = clf.cv_results_['mean_test_score']
    params = clf.cv_results_['params']
    for mean,param in zip(means,params):
        print("Score %f with: %r" % (mean, param))
    # we can query the best parameter value and its accuracy score
    print ("The best parameter value is: %r The best score is: %f" %(clf.best_params_,clf.best_score_) )
    return clf.best_params_
   
```

```python
%%time
# find the best value for the three main hyperparameters
print('The cv result for n_estimator')
parameters_n_estimators = {'n_estimators':[10,50,100,150,200,300,500]}
cv_to_choose_parameter(parameters_n_estimators)
print()
print('The cv result for max_features')
parameters_max_feature = {'max_features':[2,4,6,8,10]}
cv_to_choose_parameter(parameters_max_feature)
print()
print('The cv result for min_samples_leaf')
parameters_min_leaf={'min_samples_leaf':[2,4,8,10,12]}
cv_to_choose_parameter(parameters_min_leaf)
```

Observing all the outcomes of these methods, we consider to combine each optimal outcome from them. Therefore the n_estimators is 500, the max_features is 8 and min_samples_leaf is 4, then build the final model for it.

* max_features: The number of features to consider when looking for the best split. Increasing it can improve the accuracy of individual trees, but it will reduce the calculation speed of the model.
* n_estimators: The number of trees in the forest. More individual trees can improve the efficiency but it will reduce the calculation speed of the model.
* min_samples_leaf: The minimum number of samples required to be at a leaf node. This may have the effect of smoothing the model, especially in regression.

```python
randomState_rf_final = 1024
rf_final = RandomForestRegressor(n_estimators=500, random_state=randomState_rf, max_features = 8, min_samples_leaf = 4)
```
